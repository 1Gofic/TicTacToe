```python
from IPython.display import clear_output  # Импорт очистки вывода
import random
import pandas as pd
import numpy as np
from time import sleep

def data_load():  # Загрузка таблицы всех комбинаций с результатами
    data = pd.read_csv("Dataset_XO.csv", dtype=str)
    data = data.drop("Unnamed: 0", axis=1)
    data = data.replace(to_replace=np.nan, value=None)
    data = data.replace(to_replace=[np.nan, ""], value="")
    data = data.drop_duplicates()
    return data


# Функция отражения игровой доски
def igrovaya_doska(kto_igrok, vid_protivnika, list_hodov, list_poley):
    clear_output(wait=False)  # Очистка экрана

    print(" ")

    match vid_protivnika:  # Вывод названия для режима игры
        case "PvP":
            print("Режим игры для двух игроков.")
        case "PvE_1":
            print("Режим игры против легкого ИИ.")
        case "PvE_2":
            print("Режим игры против среднего ИИ.")
        case "PvE_3":
            print("Режим игры против сложного ИИ. Он ходит первым!")
        case "PvE_4":
            print("Режим игры против нечестного ИИ. Он знает все ходы!")
        case "EvE":
            print("Битва между ИИ!")

    print(" ")
    print(f"Сейчас ходит - {kto_igrok}. Введите номер ячейки:")
    print(f"Номера свободных мест - {list(set(list_hodov))}.")
    print(" ")

    print(" --- --- --- ")
    # Вывод игрового поля, в котором на место пропусков встают значения из списка
    print(f"| {list_poley[0]} | {list_poley[1]} | {list_poley[2]} |")
    print(" --- --- --- ")
    print(f"| {list_poley[3]} | {list_poley[4]} | {list_poley[5]} |")
    print(" --- --- --- ")
    print(f"| {list_poley[6]} | {list_poley[7]} | {list_poley[8]} |")
    print(" --- --- --- ")


def pustoy_list():  # Функция возрвращающая пустой набор значений
    return [" " for i in range(0, 9)]


def X_ili_O(nomer_hoda):  # Функция, которая показывает чей ход в зависимости от его номера
    if (nomer_hoda % 2) == 0:
        itog_igrok = "X"
    else:
        itog_igrok = "O"

    return itog_igrok


# Функция, которая оставляет в списке только свободные поля
def list_nomerov_poley(list_poley):
    nomera = pustoy_list()

    for i in range(0, 9):
        if list_poley[i] == " ":
            nomera[i] = i  # Провереяется условие на пустое место

    return nomera


def hod_igroka(list_hodov):  # Функция, которая выводит результат решения игрока
    pole_hoda = 1000

    # Принимает только значения, которые есть в списке свободных
    while not (int(pole_hoda) in list_hodov):
        try:
            pole_hoda = int(input())
        except:
            pole_hoda = 1000

    return int(pole_hoda)


def pobeda_ili_net(list_poley):  # Функция, которая проверяет наличие победителя
    resultat_proverki = False

    if list_poley[0] == list_poley[3] == list_poley[6] != " ":
        resultat_proverki = True
    if list_poley[1] == list_poley[4] == list_poley[7] != " ":
        resultat_proverki = True
    if list_poley[2] == list_poley[5] == list_poley[8] != " ":
        resultat_proverki = True
    if list_poley[0] == list_poley[1] == list_poley[2] != " ":
        resultat_proverki = True
    if list_poley[3] == list_poley[4] == list_poley[5] != " ":
        resultat_proverki = True
    if list_poley[6] == list_poley[7] == list_poley[8] != " ":
        resultat_proverki = True
    if list_poley[0] == list_poley[4] == list_poley[8] != " ":
        resultat_proverki = True
    if list_poley[2] == list_poley[4] == list_poley[6] != " ":
        resultat_proverki = True

    return resultat_proverki


def hod_ai_1(list_poley):  # Функция, возвращающая случайное число из свободных, как ход ИИ
    hod_ai = random.choice([0, 1, 2, 3, 4, 5, 6, 7, 8])
    while list_poley[hod_ai] != " ":
        hod_ai = random.choice([0, 1, 2, 3, 4, 5, 6, 7, 8])
    return hod_ai


# Функция, непосредственно проверяющая определенную комбинацию
def proverka_uslovia(pole_1, pole_2, pole_3, list_poley):
    hod_ai = 1000

    match list_poley[pole_1] + list_poley[pole_2] + list_poley[pole_3]:
        case "XX ":
            hod_ai = pole_3
        case "X X":
            hod_ai = pole_2
        case " XX":
            hod_ai = pole_1
        case "OO ":
            hod_ai = pole_3
        case "O O":
            hod_ai = pole_2
        case " OO":
            hod_ai = pole_1

    return hod_ai


def hod_ai_2(list_poley):  # Функция, в которой происходит проверка на наличие 2-х символов Х или О подряд и возвращающая ход ИИ
    hod_ai = 1000

    if hod_ai == 1000:
        # Проверяется каждый вариант
        hod_ai = proverka_uslovia(0, 3, 6, list_poley)
    if hod_ai == 1000:
        hod_ai = proverka_uslovia(1, 4, 7, list_poley)
    if hod_ai == 1000:
        hod_ai = proverka_uslovia(2, 5, 8, list_poley)
    if hod_ai == 1000:
        hod_ai = proverka_uslovia(0, 1, 2, list_poley)
    if hod_ai == 1000:
        hod_ai = proverka_uslovia(3, 4, 5, list_poley)
    if hod_ai == 1000:
        hod_ai = proverka_uslovia(6, 7, 8, list_poley)
    if hod_ai == 1000:
        hod_ai = proverka_uslovia(0, 4, 8, list_poley)
    if hod_ai == 1000:
        hod_ai = proverka_uslovia(2, 4, 6, list_poley)

    if hod_ai == 1000:
        hod_ai = random.choice([0, 1, 2, 3, 4, 5, 6, 7, 8])
        while list_poley[hod_ai] != " ":
            # Если такого наличия нет, то берётся случайное число
            hod_ai = random.choice([0, 1, 2, 3, 4, 5, 6, 7, 8])

    return hod_ai


# Поиск среди всех комбинаций тех, что приведут к победе
def hod_ai_3(df, hronologia, list_poley):

    hod_ai = 1000

    if hod_ai == 1000:
        # Проверяется каждый вариант
        hod_ai = proverka_uslovia(0, 3, 6, list_poley)
    if hod_ai == 1000:
        hod_ai = proverka_uslovia(1, 4, 7, list_poley)
    if hod_ai == 1000:
        hod_ai = proverka_uslovia(2, 5, 8, list_poley)
    if hod_ai == 1000:
        hod_ai = proverka_uslovia(0, 1, 2, list_poley)
    if hod_ai == 1000:
        hod_ai = proverka_uslovia(3, 4, 5, list_poley)
    if hod_ai == 1000:
        hod_ai = proverka_uslovia(6, 7, 8, list_poley)
    if hod_ai == 1000:
        hod_ai = proverka_uslovia(0, 4, 8, list_poley)
    if hod_ai == 1000:
        hod_ai = proverka_uslovia(2, 4, 6, list_poley)

    if hod_ai == 1000:
        nomer_hoda = len(hronologia)
        igrok = X_ili_O(nomer_hoda-1)
        filter_data = df[df["Stack"].str.startswith(hronologia)]
        filter_data = filter_data[filter_data["Win"] == igrok]
        if filter_data.shape[0] == 0:
            filter_data = df[df["Stack"].str.startswith(hronologia)]
            filter_data = filter_data[filter_data["Win"] == "-"]
        if filter_data.shape[0] == 0:
            filter_data = df[df["Stack"].str.startswith(hronologia)]
            filter_data = filter_data[filter_data["Win"] != igrok]
        filter_data['Next'] = filter_data["Stack"].str.get(nomer_hoda)
        hod_ai = random.choice([i for i in filter_data["Next"]])

    return hod_ai
```

```python
tic_tac_toe = True
df = data_load()

while tic_tac_toe:
    clear_output(wait=False)

    list_poley = pustoy_list()
    list_hodov = pustoy_list()  # Объявление переменных
    vid_protivnika = ""
    Next_step = ""
    konec = ""

    print(" ")
    print("Добро пожаловать в игру!")  # Начальный заголовок игры
    print("")
    print("Введите PvP, чтобы запустить режим для двух игроков.")
    print("Введите PvE_1, чтобы запустить режим против легкого ИИ.")
    print("Введите PvE_2, чтобы запустить режим против среднего ИИ.")
    print("Введите PvE_3, чтобы запустить режим против сложного ИИ. Он ходит первым!")
    print("Введите PvE_4, чтобы запустить режим против нечестного ИИ. Он знает все ходы!")
    print("Введите EvE, чтобы запустить режим ИИ против ИИ!")
    print(" ")

    # Получение от игрока выбора противника
    while not (vid_protivnika == "PvP" or vid_protivnika == "PvE_1" or vid_protivnika == "PvE_2" or vid_protivnika == "PvE_3" or vid_protivnika == "PvE_4" or vid_protivnika == "EvE"):
        vid_protivnika = str(input())

    for i in range(0, 10):  # Пошаговый ход игры
        list_hodov = list_nomerov_poley(list_poley)
        igrovaya_doska(X_ili_O(i), vid_protivnika, list_hodov,
                       list_poley)  # Отображение игровой доски

        if pobeda_ili_net(list_poley):
            print(" ")
            # Проверка победы и в случае наличия показ победителя
            print(f"Победа игрока - {X_ili_O(i-1)}")
            break

        if list_hodov == pustoy_list():  # Проверка на конец игры ничьей
            print(" ")
            print("Ничья")
            break

        

        # print(f"AI_1 думает сходить так - {hod_ai_1(list_poley)}.")
        # print(f"AI_2 думает сходить так - {hod_ai_2(list_poley)}.") # Показ решений, которые мог бы предпринять компьютер
        # print(f"AI_3 думает сходить так - {hod_ai_3(df,Next_step,list_poley)}.")
        print(
            f"Совет для следующего хода - {hod_ai_3(df,Next_step,list_poley)}.")

        if X_ili_O(i) == "O":
            if vid_protivnika == "PvE_1":
                hod = hod_ai_1(list_poley)
            elif vid_protivnika == "PvE_2":
                hod = hod_ai_2(list_poley)
            elif vid_protivnika == "PvE_4":
                hod = int(hod_ai_3(df, Next_step, list_poley))
            elif vid_protivnika == "EvE":
                sleep(1)
                hod = int(hod_ai_3(df, Next_step, list_poley))
            else:
                hod = hod_igroka(list_hodov)
        else:
            if vid_protivnika == "PvE_3":
                hod = hod_ai_2(list_poley)
            elif vid_protivnika == "EvE":
                sleep(1)
                hod = int(hod_ai_3(df, Next_step, list_poley))
            else:
                hod = hod_igroka(list_hodov)

        Next_step = Next_step + str(hod)
        list_poley[hod] = X_ili_O(i)

    print("")
    print("Поздравляю!")
    print("Игра окончена, хотите повторить?..")
    print("Введите 'Yes' или 'No'.")
    print("")

    while not (konec == "Yes" or konec == "No"):  # Получение ответа от игрока
        konec = str(input())

    if konec == "No":
        tic_tac_toe = False
```
